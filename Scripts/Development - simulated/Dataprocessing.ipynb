### Lav timeværdier og timestamps med datetime 

```python
# Afrund til laveste hele nummer i første kollone

df_fault.iloc[:, 0] = np.floor(df_fault.iloc[:, 0])

#df_faultfree.iloc[:, 0] = np.floor(df_faultfree.iloc[:, 0])

df_fault
```

```python
# Behold kun 1 af hver måling i hver time 

#df_faultfree = df_faultfree.drop_duplicates(subset=df_faultfree.columns[0])

df_fault = df_fault.drop_duplicates(subset=df_fault.columns[0])

df_fault
```

```python
### Make time and date stamps on the data 

# Define the starting point (e.g., '2023-01-01 00:00:00')
start_time = pd.Timestamp('2023-01-01 00:00:00')

# Convert numbers to timestamps assuming they are hours since the start_time
df_fault.iloc[:, 0] = start_time + pd.to_timedelta(df_fault.iloc[:, 0], unit='h')

df_fault = df_fault.reset_index(drop=True)

# Convert numbers to timestamps assuming they are hours since the start_time
#df_faultfree.iloc[:, 0] = start_time + pd.to_timedelta(df_faultfree.iloc[:, 0], unit='h')

#df_faultfree = df_faultfree.reset_index(drop=True)

df_fault
```

### Make sure there is zeros in all FDD columns 

```python
df_fault.loc[:, 'FDD, zone 1'] = 0
#df_fault.loc[:, 'FDD, zone 5'] = 0
df_fault.loc[:, 'FDD, zone 8'] = 0

#df_faultfree.loc[:, 'FDD, zone 1'] = 0
#df_faultfree.loc[:, 'FDD, zone 5'] = 0
#df_faultfree.loc[:, 'FDD, zone 8'] = 0
```

## Tilføj vejrdata (Temp + GHI)


```python
df_weather

df_weather["Time(utc)"] = pd.to_datetime(df_weather["Time(utc)"])

# Resample to hourly average
df_weather = df_weather.resample("H", on="Time(utc)").mean()
```

```python
# Ensure column length matches df_target
column_to_add1 = df_weather['GHI']
column_to_add2 = df_weather['air_temperature']

# Add to df_target
df_fault.insert(1, 'Outdoor air temp', column_to_add2.values)

df_fault.insert(1, 'Global Horizontal Irradiance, W/m2', column_to_add1.values)

# if Nan, fill with 0
df_fault["Global Horizontal Irradiance, W/m2"] = df_fault["Global Horizontal Irradiance, W/m2"].fillna(0)

df_fault
```

## Normaliser varmedata 


```python
# Find the max value 
max_value_zone1 = df_fault['Local heating units, W, zone 1'].max()

max_value_zone8 = df_fault['Local heating units, W, zone 8'].max()

max_value_building = df_fault['Ideal heaters and other local u'].max()

# Apply division only where the value is not 0
df_fault['Local heating units, W, zone 1'] = np.where(df_fault['Local heating units, W, zone 1'] == 0, 0, df_fault['Local heating units, W, zone 1']/max_value_zone1)

df_fault['Local heating units, W, zone 8'] = np.where(df_fault['Local heating units, W, zone 8'] == 0, 0, df_fault['Local heating units, W, zone 8']/max_value_zone8)

df_fault['Ideal heaters and other local u'] = np.where(df_fault['Ideal heaters and other local u'] == 0, 0, df_fault['Ideal heaters and other local u']/max_value_building)


df_fault
```
